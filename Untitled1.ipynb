```python
def read_files(cur_dir):
    data_folder = os.path.join(cur_dir, 'project-data')
    tar_path = os.path.join(data_folder + '.tar.gz')
    with tarfile.open(tar_path, mode='r:gz') as tar:
        tar.extractall(path=cur_dir)
    data_folder_extract = os.path.join(cur_dir,'project-data/images')
    train_images = os.path.join(data_folder_extract,'train')
    annotations = os.path.join(cur_dir,'project-data/annotations/train')
    return train_images,annotations
```

```python
import xml.etree.ElementTree as ET

def parse_file(filename):
    """ Parse a PASCAL VOC xml file """
    tree = ET.parse(filename)
    objects = []
    for obj in tree.findall('object'):
        obj_struct = {}
        obj_struct['name'] = obj.find('name').text
        bbox = obj.find('bndbox')
        obj_struct['bbox'] = [int(float(bbox.find('xmin').text)),
                              int(float(bbox.find('ymin').text)),
                              int(float(bbox.find('xmax').text))-int(float(bbox.find('xmin').text)),
                              int(float(bbox.find('ymax').text))-int(float(bbox.find('ymin').text))]
        objects.append(obj_struct)

    return objects
```

```python
def annots_to_coords(annots):
    coords = []
    for an in annots:
        coords.append(np.array(an['bbox']))
    
    return np.array(coords)
```

```python
# A rectangle is defined by the coordinates of its upper left and lower right corners
def check_intersection(ul1,tr1,ul2,tr2):
    minx1,miny1 = ul1
    maxx1,maxy1  = tr1
    minx2,miny2 = ul2
    maxx2,maxy2 = tr2
    return (minx2 <= maxx1 and maxx2 >= minx1 and miny2 <= maxy1 and maxy2 >= miny1)
```

```python
def wiener_filt(img):
    psf = np.ones((5, 5)) / 25
    im = convolve2d(color.rgb2gray(img), psf, 'same')
    im += 0.1 * im.std() * np.random.standard_normal(im.shape)
    denoised_im = skimage.restoration.wiener(im, psf, 1100)
    return denoised_im
```

```python
X1 = []
for pos_im in train_imgs_pos:
    cur_im = skimage.io.imread(os.path.join(train_path_pos,pos_im))
    #tmp = skimage.restoration.denoise_bilateral(cur_im)
    tmp = skimage.filters.sobel(color.rgb2gray(cur_im))
    cur_im = transform.resize(cur_im,(50,50,3))
    X1.append(cur_im)
num_positives = len(X1)
for neg_im in train_imgs_neg:
    cur_im = skimage.io.imread(os.path.join(train_path_neg,neg_im))
    #tmp = skimage.restoration.denoise_bilateral(cur_im)
    tmp = skimage.filters.sobel(color.rgb2gray(cur_im))
    #if tmp.shape[0] != 50 or tmp.shape[1] != 50:
        #continue
    
    X1.append(cur_im)
    #if len(X1) - num_positives == num_positives:
        #break
num_negatives = 70000 - num_positives
```

```python
X1 = X1[:30000]
num_negatives = len(X1) - num_positives
wiener_arr = list(map(lambda el:wiener_filt(el),X1))
wiener_arr_train = np.array(wiener_arr)
wiener_arr_train = wiener_arr_train.reshape((wiener_arr_train.shape[0],50,50,1))
y_train = np.vstack((np.ones((num_positives,1)),np.zeros((num_negatives,1))))
```

```python
# Network on the plain images
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K


# dimensions of our images.
img_width, img_height = 50,50

epochs = 10
batch_size = 10

if K.image_data_format() == 'channels_first':
    input_shape = (1, img_width, img_height)
else:
    input_shape = (img_width, img_height, 1)

model1 = Sequential()
model1.add(Conv2D(32, (3, 3), input_shape=input_shape))
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))

model1.add(Conv2D(32, (3, 3)))
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))

model1.add(Conv2D(64, (3, 3)))
model1.add(Activation('relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))

model1.add(Flatten())
model1.add(Dense(64))
model1.add(Activation('relu'))
model1.add(Dropout(0.5))
model1.add(Dense(1))
model1.add(Activation('sigmoid'))

model1.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              #optimizer = 'Adam',
              metrics=['accuracy'])

model1.fit(wiener_arr_train,y_train,epochs = 5)

#model.save_weights('m1.h5')
```

```python
def sliding_window(image, stepSize, windowSize):
    # slide a window across the image
    for y in range(0, image.shape[0], stepSize):
        for x in range(0, image.shape[1], stepSize):
            # yield the current window
            yield (x, y, image[y:y + windowSize[1], x:x + windowSize[0]])
```

```python
def find_intersection(bboxA,bboxB):
    """  
    Find area of intersection of two bounding boxes
    Parameters: 
    bboxA (list): first bounding box
    bboxB (list): second bounding box
  
    Returns: 
    int: Area of intersection of bounding boxes
  
    """
    dif1 = min(bboxA[2],bboxB[2]) - max(bboxA[0],bboxB[0])
    dif2 = min(bboxA[3],bboxB[3]) - max(bboxA[1],bboxB[1])
    if dif1 <= 0 or dif2 <= 0:
        return 0
    else:
        return dif1*dif2
```

```python
def find_union(bboxA,bboxB):
    """  
    Find area of union of two bounding boxes
    Parameters:
    bboxA (list): first bounding box
    bboxB (list): second bounding box
  
    Returns: 
    int: Area of union of bounding boxes as sum of areas - area of intersection
  
    """
    area1 = (bboxA[2]-bboxA[0])*(bboxA[3] - bboxA[1])  
    area2 = (bboxB[2]-bboxB[0])*(bboxB[3] - bboxB[1]) 
    return area1 + area2  - find_intersection(bboxA,bboxB)
```
